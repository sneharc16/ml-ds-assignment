# 02 · Fair-Price Model

Train/evaluate the fair-price model, inspect the conformalized interval and SHAP.

> Synthetic data — methodology demo, not CARS24 results. Run `make pipeline-small` first so artifacts exist.

```python
from driveintent.config import load_config
from driveintent.models import price
cfg = load_config()
res = price.train(cfg)
cb = res['catboost']
print('MAE', round(cb['mae']), 'R2', round(cb['r2'],3),
      'P10-P90 coverage', round(cb['p10_p90_coverage'],3))
print('baselines:', {k: round(v['mae']) for k,v in res.items()
                      if k.startswith('baseline')})
```

## Predict + explain a single car

```python
import pandas as pd
ds = pd.read_parquet(cfg.processed_data/'price_dataset.parquet')
row = ds.iloc[0].to_dict()
print(price.predict(cfg, row))
price.explain(cfg, row, top_k=6)
```

## Predicted vs actual

```python
import plotly.express as px
d = pd.read_parquet(cfg.artifacts/'metrics'/'price_test_predictions.parquet')
px.scatter(d, x='target_price', y='predicted', opacity=0.5,
           title='Predicted vs actual (test)').show()
```
